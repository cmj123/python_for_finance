## Monte Carlo Simulation

### Import key libraries, classes and functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Setting parameters

```python
rev_m = 170
rev_stdev = 20
iterations = 1000
```

### Get potential revenuem

```python
rev = np.random.normal(rev_m, rev_stdev, iterations)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(rev)
plt.show()
```

```python
COGS = - (rev * np.random.normal(0.6,0.1))

plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

### Gross Profit

```python
gross_profit = rev + COGS

plt.figure(figsize=(15,6))
plt.plot(gross_profit)
plt.show()
```

```python
max(gross_profit)
```

```python
min(gross_profit)
```

```python
gross_profit.mean()
```

```python
gross_profit.std()
```
